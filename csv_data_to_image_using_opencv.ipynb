```python
#importing different libraries
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
image_path='/content/drive/MyDrive/image_processing_lab/lena_dark.png'
```

```python
image_read=cv2.imread(image_path)
image_gray=cv2.cvtColor(image_read, cv2.COLOR_BGR2GRAY) #converting the image to gray scale

image_array= np.array(image_gray)
```

```python
#write numpy array to csv file --> image_array_data
#newline='' -->  no extra newline characters are added

with open('/content/drive/MyDrive/image_array_data.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)
  for row in image_array:
    writer.writerow(row)
```

```python
#path of the csv filed saved
csv_file_path='/content/drive/MyDrive/image_array_data.csv'
```

```python
#Read the data from the CSV file into a NumPy array
image_array_read = np.genfromtxt(csv_file_path, delimiter=',')

# Convert the NumPy array back to an image
#cv2.convertScaleAbs --> function in OpenCV used for scaling and converting the input array to an unsigned 8-bit integer array
recreated_image = cv2.convertScaleAbs(image_array_read)

```

```python
# Display the original and recreated images
plt.figure(figsize=(10, 10))

plt.subplot(1, 2, 1)
plt.imshow(image_array, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(recreated_image, cmap='gray')
plt.title('Recreated Image')

plt.show()
```
